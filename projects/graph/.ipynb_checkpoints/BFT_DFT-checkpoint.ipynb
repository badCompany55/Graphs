## Notes and Solutions BFT and DFT Seach

### Classes for Queue and Stack

1. ### Queue
    * Queue is first in first out
        * The first oldest value in the queue is the next one out

```python
class Queue():
    # create storage component on initialization
    
    def __init__(self):
        self.queue = [] 
        
    # add values to the end of the queue
    
    def enqueue(self, value):
        self.queue.append(value)
        
    # remove value from the beginning of the queue
    
    def dequeue(self):
        if self.size() > 0:
            return self.queue.pop(0)
        else:
            return None
        
    # size method that returns the length
    
    def size(self):
        return len(self.queue)
```

2. ### Stack
    * Stack is last in first out
        * The last value in is the first value out


```python
class Stack():
    # Create storage component on initialization
    
    def __init__(self):
        self.stack = []
        
    # add value to the end of the stack
    
    def push(self, value):
        self.stack.append(value)
        
    # remove value from the end of the 
    
    def pop(self):
        if self.size() > 0:
            return self.stack.pop()
        else:
            return None
        
    # size method that returns the length
    
    def size(self):
        return len(self.stack)
    
```

3. ### Graph class implementation
    * This will contain all the methods needed for the graph. This included the following
        1. add_vertex __method to add data point to the graph__
        2. add_edge __adds path/connection between points__
        3. bft __bredth first implementation__
        4. dft __depth first implementation__
        5. dft recursive __recursive implementation__
        6. bfs __bredth first search returning the shortest path__
        7. dfs __depth first search returning the shortest path__

```python
class Graph():
    
    # storage component to store the data points
    
    def __init__(self):
        self.vertices = {}
        
    # add a vertex, will use sets to store them. Better lookup functionality and speed
    
    def add_vertex(self, vertex):
        self.vertices[vertex] = set()
        
    # add a connection between two vertices
    
    def add_edge(self, v1, v2):
        
        # select the vertex from vertices("object")
        # use the .add method to add a value to a set
        self.vertices[v1].add(v2)
        
    # bredth first implementation
    
    def bft(self, starting_vertex):
        
        # bft is first in first out.
        # the oldest value is the next one out
        # Queue is well suited for this
        
        #initialize the Queue
        
        q = Queue()
        
        # Keep track of the data points/nodes/vertices/ that have been visted
        # goint to use a set for this
        visited = set()
        
        # add the starting_vertex to the queue to start it off
        q.enqueue(starting_vertex)
        
        # get the size for use in a while loop
        size = q.size()
        
        # a set orders the values
        # if you want to see the order of how the nodes were visited, put them in a list
        return_val = []
        
        # Perform said operations while there is a value in the Queue
        while size > 0:
            # set a value the value from dequeue
            current_node = q.dequeue()
            
            # check to see if the current_node value has been visted yet
            if current_node not in visited:
                # add the current_node to visited
                visited.add(current_node)
                return_val.append(current_node)
                
                #get the neighbors of the current_node
                edges = self.vertices[current_node]
                
                #loop through neighbors and add them to the queue for processing in the while loop
                for e in edges:
                    q.enqueue(e)
                    
            # set new size of the queue
            size = q.size()
            
        return return_val
        
    # Depth first traversal
    def dft(self, starting_vertex):
    
    # depth first is first in last out
    # this makes a stack well suited for the task
    
        s = Stack()

        # use a set to keep track of visited data points/ nodes/ vertices

        visited = set()

        # initiate the stack by adding the starting_vertex to it

        s.push(starting_vertex)

        # get the size of the stack for use in while loop

        size = s.size()

        # set returns everything in order.
        # to see which vertex was visited in which order, use a list

        return_val = []

        # create a while loop that runs while the stack has a length

        while size > 0:

            # set a value to the current value of s.pop(). 
            # this is the value that we are checking against

            current_node = s.pop()

            # check to see if the current_node has been visited yet

            if current_node not in visited:

                # add node to visited
                # add node to return val

                visited.add(current_node)
                return_val.append(current_node)

                # get all of the neighbors of that current node

                edges = self.vertices[current_node]

                #loop through the neighbors pushing them to the stack to be checked

                for e in edges:
                    s.push(e)

            # change the size to reflect the the current size of the stack
            size = s.size()

        return return_val
    
    # Dft recursivly
    # Only Dft's can be implemented recursively
    
    def dft_recursive(self, starting_vertex, return_val=[]):
        # add the starting vertex to the return_val array
        return_val += [starting_vertex]
        
        # get the neighbors
        edges = self.vertices[starting_vertex]
        
        # loop through the neighbors
        for e in edges:
            
            # check if neighbor has beeen visted/return_val
            if e not in return_val:
                
                # call dft_recursive and add e to the return_val
                return_val = self.dft_recursive(e, return_val)
        
        return return_val
    
    # Find the shortest path between nodes using bft
    
    def bfts(self, starting_vertex, destination_vertex):
        q = Queue()
        visited = set()
        
        '''
        Unlike a regular travseral, in order to find the shortest path
        each variation of the whole path has to be stored.
        In order to do this the path is stored in a list
        '''
        
        # initialize the Queue with the starting vertex in an array
        
        q.enqueue([starting_vertex])
        size = q.size()
        
        while size > 0:
            # set the current node to the value of the dequeue
            # this will be an array
            current_node = q.dequeue()
            
            # grab the value of the last element of the current_node
            # this will be used to check if the destination has been reached
            
            single_node = current_node[-1]
            
            # check if single node has been visited
            
            if single_node not in visited:
                # get the neighbors of the single_node
                
                edges = self.vertices[single_node]
                
                # loop through the neighbors
                
                for e in edges:
                    # create a copy of the current_node array
                    # append the neighbor to the array
                    # pass that entire array to the queue
                        # this is how each variation of the path is checked and tracked
                    new_path = current_node[:]
                    new_path.append(e)
                    q.enqueue(new_path)
                    
                    # check or e being the destination_vertex
                    # this would mean that we have found the shortest path the the destination
                    
                    if e is destination_vertex:
                        return new_path
                    
                # if none of the edges equal the destination we need to move to the next possible vertex
                # add the visted vertex to visited and repeat
                
                visited.add(single_node)
                
    # find the shortes path between nodes using dft
    
    def dfts(self, starting_vertex, destination_vertex):
        
        s = Stack()
        visited = set()
        s.push([starting_vertex])
        size = s.size()
        
        while size > 0:
            current_node = s.pop()
            single_node = current_node[-1]
            
            if single_node not in visited:
                edges = self.vertices[single_node]
                
                for e in edges:
                    new_path = current_node[:]
                    new_path.append(e)
                    s.push(new_path)
                    
                    if e is destination_vertex:
                        return new_path
                    
                visited.add(single_node)
```

```python

graph = Graph()

graph.add_vertex(1)
graph.add_vertex(2)
graph.add_vertex(3)
graph.add_vertex(4)
graph.add_vertex(5)
graph.add_vertex(6)

graph.add_vertex(7)
graph.add_edge(5, 3)
graph.add_edge(6, 3)
graph.add_edge(7, 1)
graph.add_edge(4, 7)
graph.add_edge(1, 2)
graph.add_edge(7, 6)
graph.add_edge(2, 4)
graph.add_edge(3, 5)
graph.add_edge(2, 3)
graph.add_edge(4, 6)


print("vertices", graph.vertices)

print("Bredth First Transversal", graph.bft(1))
print("Depth First Transversal", graph.dft(1))

print("Depth First Transveral Recursively", graph.dft_recursive(1))
print("Bredth First Transversal, Shortest Path", graph.bfts(1,6))

print("Depth First Transveral, Shortest Path", graph.dfts(1,6) )
```
